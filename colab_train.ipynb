# DroneWaste Colab Training

This notebook clones the project, installs the selected training dependencies, downloads the DroneWaste Zenodo dataset, and starts a Colab-friendly training run.

```python
from pathlib import Path
import os
import subprocess

REPO_URL = "https://github.com/Leo890728/dronewaste.git"  #@param {type:"string"}
PROJECT_DIR = "/content/dronewaste"  #@param {type:"string"}

project_dir = Path(PROJECT_DIR)
if not (project_dir / ".git").exists():
    subprocess.run(["git", "clone", "--recursive", REPO_URL, PROJECT_DIR], check=True)

os.chdir(PROJECT_DIR)
print(f"Working directory: {Path.cwd()}")
```

```python
import os
import subprocess

ARCH = "yolov8"  #@param ["yolov8", "yolov12", "faster"]
MODEL = "yolov8n"  #@param {type:"string"}
SITE_INDICES = "0"  #@param {type:"string"}
EPOCHS = 10  #@param {type:"integer"}
PATIENCE = 5  #@param {type:"integer"}
BATCH_SIZE = 16  #@param {type:"integer"}
IMG_SIZE = 640  #@param {type:"integer"}
STORAGE = "/content"  #@param {type:"string"}
AUTO_INSTALL = True  #@param {type:"boolean"}
AUTO_DOWNLOAD = True  #@param {type:"boolean"}

env = os.environ.copy()
env.update(
    {
        "ARCH": ARCH,
        "MODEL": MODEL,
        "SITE_INDICES": SITE_INDICES,
        "EPOCHS": str(EPOCHS),
        "PATIENCE": str(PATIENCE),
        "BATCH_SIZE": str(BATCH_SIZE),
        "IMG_SIZE": str(IMG_SIZE),
        "STORAGE": STORAGE,
        "AUTO_INSTALL": "1" if AUTO_INSTALL else "0",
        "AUTO_DOWNLOAD": "1" if AUTO_DOWNLOAD else "0",
        "WANDB_MODE": "offline",
    }
)

subprocess.run(["bash", "training/colab_train.sh"], check=True, env=env)
```
